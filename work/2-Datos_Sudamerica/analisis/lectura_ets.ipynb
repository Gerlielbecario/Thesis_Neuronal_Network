```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Deberia haberlos importado, pero bueno son los mismos que use, por ahi lo corro de nuevo
Umbrales = [0.01,0.5,1,3,5,10,15,20,25,50,75,100,120,130,140,150,160,170,180,190,200,220,250,270,300,330,350,390,400,450,500,550,600,650,700,750,800,850,900,1000,1100,1200]


#Path ETS Zona tropical
path_ets_tropical = '/home/fernando.huaranca/datosmunin/salida_ets/ets_tropical.npz'
tropical = np.load(path_ets_tropical)
tropical_ets = tropical['ETS']

#Path ETS Zona Lat medias
path_ets_medios = '/home/fernando.huaranca/datosmunin/salida_ets/ets_medios.npz'
medios = np.load(path_ets_medios)
medios_ets = medios['ETS']
```

```python
fig,axs = plt.subplots(figsize=(12,6),nrows=1,ncols=2)
fig.suptitle('Equitable Threat Score (ETS)')

#Grafico de Umbrales Completos (Figura izquierda)
axs[0].plot(Umbrales,tropical_ets,label='ETS Latitudes Tropicales')
axs[0].plot(Umbrales,medios_ets,label='ETS Latitudes Medias')
axs[0].set_title('Umbrales hasta 1000 mm')
axs[0].legend()
axs[0].grid(True)
#Grafico de Umbrales hasta 20

axs[1].plot(Umbrales[0:20],tropical_ets[0:20],label='ETS Latitudes Tropicales')
axs[1].plot(Umbrales[0:20],medios_ets[0:20],label='ETS Latitudes Medias')
axs[1].set_title('Umbrales hasta 175 mm')
axs[1].legend()
axs[1].grid(True)
plt.savefig('/home/fernando.huaranca/datosmunin2/Thesis_Neuronal_Network/work/2-Datos_Sudamerica/salidas/grafico_ets.png',dpi=300)
```
