```python
import numpy as np
import xesmf as xe

# Cargar tus datos desde el archivo NPZ
data_npz = np.load('/home/fernando.huaranca/datosmunin3/Gsmap_24hs/Gsmap_R0.1_24hs_2000-01-04.npz')

# Extraer las variables de longitud y latitud
lon_in = data_npz['longitudes']
lat_in = data_npz['latitudes']
data_in = data_npz['pp_daily']  # Ajustar el nombre segn tus datos

# Cargar el conjunto de datos de salida desde otro archivo NPZ
data_out_npz = np.load('/home/fernando.huaranca/datosmunin3/GFS_24hs/GFS_R0.25_24hs_2000-01-04.npz')
lon_out = data_out_npz['longitudes']
lat_out = data_out_npz['latitudes']

# Crear las cuadrculas de entrada y salida con xesmf
grid_in = xe.grid({'lon': lon_in, 'lat': lat_in})
grid_out = xe.grid({'lon': lon_out, 'lat': lat_out})

# Crear el Regridder
regridder = xe.Regridder(grid_in, grid_out, 'conservative')

# Realizar la regridificacin
resultado = regridder(data_in)
```
