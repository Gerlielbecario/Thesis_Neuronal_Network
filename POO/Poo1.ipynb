# Programacion Orientada a Objetos

¿ En que consiste?

* Trasladar la naturaleza de los objetos de la vida real al código de programación


¿Cuál es la naturaleza de un objeto en la vida real? 

* Los objetos tienen un estado, un comportamiento (¿Que pueden hacer?) y unas propiedades

Pongamos un ejemplo: El objeto Coche

* ¿Cuál es el estado de un objeto coche? Un coche puede estar parado, circulando, aparcado, etc

* ¿Que propiedades tiene un coche? Un coche tiene un color, un peso, un tamaño, etc

* ¿ Que comportamiento tiene un coche? Un coche puede arrancar,frenar, acelerar, girar, etc





## Ventajas

* Programas dividos en trozos, partes, modulos o clases. Modularización
* Muy reutilizable. Herencia
* Si existe fallo en alguna linea de código, el programa continuará con su funcionamiento. Tratamiento de Excepciones.
* Encapsulamiento

## Clase

Un coche:

* El chasis
* Las 4 ruedas

## Modularización

Lo normal al crear una apliación compleja es que este formada por varias clases. 

El concepto de modularización deriva de que una aplicación puede estar compuesta de varias clases

Cada uno de estos elementos o módulos funcionan de forma diferente. 

Si se estropea una parte se puede reparar solo la parte fallada.

## Encapsulación

El funcionamiento interno de un objeto nada sabe y esta protegido del exterior


## MEtodos de acceso

Para conectar las diferentes clases entre si, para que funcionen como una clave o un equipo

## Nomenclatura del punto

Para acceder a las diferentes clases

* Objeto:
    * Accediendo a propiedades de objeto desde código
        * miCoche.color = 'rojo'
        * miCoche.peso = 1500
        * miCoche.ancho = 2000
        * miCoche.alto = 900
    
    * Accediendo a comportamiento de objeto desde código 
        * miCoche.arranca()
        * miCoche.frena()
        * miCoche.gira()
        * miCoche.acelera()
